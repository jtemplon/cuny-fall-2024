```python
# just plugged it in
most_btc = explorer.get_address('1HrPF5CPqJiWbkroxheU5LcHL7bZNDi76v')
```

```python
transactions = most_btc.transactions
```

```python
transactions
```

```python
input_addresses = {}
output_addresses = {}

for t in transactions:
    for i in t.inputs:
        if i.address in input_addresses:
            input_addresses[i.address] += i.value
        else:
            input_addresses[i.address] = i.value
    for o in t.outputs:
        if o.address in output_addresses:
            output_addresses[o.address] += o.value
        else:
            output_addresses[o.address] = o.value
```

```python
inputs_df = pd.DataFrame.from_dict(input_addresses, orient='index', columns=['input_total'])
outputs_df = pd.DataFrame.from_dict(output_addresses, orient='index', columns=['output_total'])
```

```python
total_transactions = (
    inputs_df
    .join(outputs_df, how="outer")
    .fillna(0)
    .assign(total_transacted = lambda df: df.sum(axis=1))
    .sort_values("total_transacted", ascending=False)
)
```

```python
# it didn't interact with any of the others?
total_transactions.reset_index().loc[lambda x: x["index"].isin([ a[0] for a in addresses ])]
```

```python
# six addresses transacted in and out with the address
total_transactions.loc[lambda x: (x["input_total"] > 0) & (x["output_total"] > 0)]
```

```python
def convert_satoshis_to_usd(num):
    btc = num * 0.00000001 # there are 100 million satoshis in a bitcoin
    usd = btc * 62922.60 # price as of 9/26 would be better to do on day of transaction
    return usd
```

```python
total_transactions["input_$"] = total_transactions["input_total"].apply(lambda x: convert_satoshis_to_usd(x))
total_transactions["output_$"] = total_transactions["output_total"].apply(lambda x: convert_satoshis_to_usd(x))
total_transactions["total_$"] = total_transactions["total_transacted"].apply(lambda x: convert_satoshis_to_usd(x))
```
